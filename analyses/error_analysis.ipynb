```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("./take1.csv")
```

```python
fig, axes = plt.subplots(nrows=3, figsize=(10, 5), sharex=True, sharey=True)

for take in (1,2,3):
    df = pd.read_csv(f"./take{take}.csv")
    df.iloc[:,1:].diff().boxplot(ax=axes[take-1], showfliers=False)
    axes[take-1].set_ylabel(f"Take {take}")
plt.tight_layout()
plt.show()
fig.savefig('3-takes-errors.png')
```
